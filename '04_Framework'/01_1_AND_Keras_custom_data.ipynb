* 1학기때 SVM 을 이용해서 분류했던 예를 신경망 구조로 해결하는 방법입니다.

```python
# tensorflow와 tf.keras를 임포트합니다
import tensorflow as tf
from tensorflow import keras
```

```python
tf.__version__
```

```python
# 헬퍼(helper) 라이브러리를 임포트합니다
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.array([[1, 2], [1, 5], [2, 2], [3, 3.5], [7, 5], [9, 4], [8, 2]]) #training_points
y_data = [1, 1, 1, 1, 0, 0, 0]

x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)   ##전 파일과는 다르게 데이터의 구조(행렬)를 변경한 것을 확인
```

```python
plt.scatter(x_data[:, 0], x_data[:, 1], c=y_data)
```

```python
from tensorflow.keras import layers
from tensorflow.keras import activations
from tensorflow.keras import optimizers
from tensorflow.keras import models
```

```python
model = models.Sequential()
model.add(layers.Dense(2, input_dim=2))
model.add(layers.Activation('tanh'))
model.add(layers.Dense(1))
model.add(layers.Activation('sigmoid'))
```

```python
sgd = optimizers.SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
model.summary()   ##모델 sequential에 대한 다양한 정보와 결과 모양을 확인
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model_and.png', show_shapes=True)
```

```python
history = model.fit(x_data, y_data,  batch_size=1, epochs=500)    ##x,y데이터에 대해 배치사이즈를 1로하고 에포크를 500으로 했을 때의 결과
```

### 학습 결과 그려보기

```python
plt.plot(history.history['loss'])
plt.title('Loss')
plt.xlabel('epoch')   ##loss함수에 대한 출력결과
```

```python
plt.plot(history.history['accuracy'])
plt.title('Training accuracy')
plt.xlabel('epoch')   ##학습 정확도에 대한 그래프결과
```

###  확률 찍어보기

```python
hypothesis = model.predict(x_data)
print(hypothesis)   ##x데이터에 대한 확률 출력
```

```python
predicted = hypothesis > 0.5
print(predicted)     ##가중치를 0.5이상으로 했을 때의 예측치 결과 확인
```
